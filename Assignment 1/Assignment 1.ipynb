# Question 1: Softmax

```python
import tensorflow as tf
from tensorflow.keras import layers, models
import matplotlib.pyplot as plt
import numpy as np
from keras.callbacks import Callback
import keras
```

```python
mnist = tf.keras.datasets.mnist
(train_images, train_labels), (test_images, test_labels) = mnist.load_data()

train_images = train_images.reshape((60000, 28, 28, 1))
test_images = test_images.reshape((10000, 28, 28, 1))

train_images, test_images = train_images / 255.0, test_images / 255.0
```

```python
train_labels = tf.keras.utils.to_categorical(train_labels, 10)
test_labels = tf.keras.utils.to_categorical(test_labels, 10)
```

```python
model = models.Sequential([layers.Flatten(input_shape=(28, 28, 1)),layers.Dense(10, activation='softmax')])
```

```python
model.summary()
```

```python
model.compile(optimizer='sgd',loss='categorical_crossentropy',metrics=['accuracy'])
```

```python
def view_weights(model):
    weights, biases = model.layers[1].get_weights()
    weights = weights.reshape(28, 28, 10)
    fig, axes = plt.subplots(1, 10, figsize=(16, 4))
    for i in range(10):
        axes[i].imshow(weights[:, :, i], cmap='viridis')
        axes[i].axis('off')
    plt.show()
```

```python
l=[]
```

```python
for epoch in range(200):
    history = model.fit(train_images, train_labels, epochs=1,batch_size = 120, verbose=1)
    test_loss, test_acc = model.evaluate(test_images, test_labels, verbose=0)
    print(f"Epoch {epoch + 1}: Test accuracy: {test_acc * 100:.2f}%")
    l.append(test_acc * 100)
    if (epoch ==19):
        view_weights(model)
```

# Question 2 : Naive Autoencoder

```python
import keras
from keras import layers
from keras.datasets import mnist
```

```python
encoding_dim = 100 
```

```python
input_img = keras.Input(shape=(784,))
encoded = layers.Dense(encoding_dim, activation='sigmoid')(input_img)
decoded = layers.Dense(784)(encoded)
```

```python
autoencoder = keras.Model(input_img, decoded)
```

```python
encoder = keras.Model(input_img, encoded)
```

```python
encoded_input = keras.Input(shape=(encoding_dim,))
decoder_layer = autoencoder.layers[-1]
```

```python
decoder = keras.Model(encoded_input, decoder_layer(encoded_input))
```

```python
opt = keras.optimizers.SGD(learning_rate=0.8)

autoencoder.compile(optimizer=opt, loss='mse')
```

```python
autoencoder.summary()
```

# Question 2b Sparse AutoEncoder

```python
import numpy as np
import tensorflow as tf
from tensorflow.keras import backend as K
import keras
from keras.datasets import mnist
import matplotlib.pyplot as plt
from tensorflow.keras.layers import Input, Dense, Flatten
from tensorflow.keras.models import Model
from tensorflow.keras.optimizers import SGD 
from tensorflow.keras import regularizers
from tensorflow.keras.models import Sequential
```

```python
(x_train, y_train), (x_test, y_test) = mnist.load_data()
x_train = x_train/255.0
x_test = x_test/ 255.0
```

```python
print(x_train.shape, x_test.shape)
```

```python
x_train = x_train.reshape((len(x_train), np.prod(x_train.shape[1:])))
x_test = x_test.reshape((len(x_test), np.prod(x_test.shape[1:])))
```

```python
print(x_train.shape, x_test.shape)
```

```python
def sparse_regularizer(activation_matrix):
    alpha = 0.2
    beta = 2
    alpha_hat = tf.reduce_mean(activation_matrix) 
    KL_divergence = alpha*(tf.math.log(alpha/alpha_hat)) + (1-alpha)*(K.log((1-alpha)/(1-alpha_hat)))
    sum_kl = K.sum(KL_divergence) 
    return beta * sum_kl
```

```python
input_layer = Input(shape=(784,))
encoded = Dense(100, activation='sigmoid',activity_regularizer=sparse_regularizer)(input_layer)  # Encoder layer
decoded = Dense(784)(encoded)  # Decoder layer
```

```python
autoencoder2 = Model(input_layer, decoded)
```

```python
autoencoder2.summary()
```

```python
sgd = SGD(learning_rate=0.4)
autoencoder2.compile(optimizer=sgd, loss='mse')
```

```python
encoder = keras.Model(input_layer, encoded)
```

```python
autoencoder2.fit(x_train, x_train, epochs=500, batch_size=150, validation_data=(x_test, x_test))
```

```python
test_loss = autoencoder2.evaluate(x_test, x_test)
print(f"Test loss: {test_loss:.4f}")
```

```python
encoder_weights = autoencoder2.layers[1].get_weights()[0]
plt.figure(figsize=(10, 10))
for i in range(24):
    ax = plt.subplot(6, 4, i + 1)
    ax.imshow(encoder_weights[:, i].reshape(28, 28), cmap='viridis')
    plt.axis('off')
plt.suptitle('Encoder Weights')
plt.show()
```

```python
encoded_images = (encoder.predict(x_test))
```

```python
plt.figure(figsize=(10, 10))
plt.pcolormesh(encoded_images[:150], cmap='viridis')
plt.colorbar(label='Color Scale')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.gca().invert_yaxis()
plt.title('Color Mesh Plot')
plt.show()
```

```python
n = 10 
decoded_images = autoencoder2.predict(x_test[:n])

plt.figure(figsize=(20, 4))
for i in range(n):
    ax = plt.subplot(2, n, i + 1)
    plt.imshow(x_test[i].reshape(28, 28), cmap='viridis')
    plt.axis('off')

    ax = plt.subplot(2, n, i + 1 + n)
    plt.imshow(decoded_images[i].reshape(28, 28), cmap='viridis')
    plt.axis('off')

plt.suptitle('Original and Reconstructed Images')
plt.show()
```

```python
(x_train, y_train), (x_test, y_test) = mnist.load_data()
x_train = x_train/255.0
x_test = x_test/ 255.0
```

```python
model2 = Sequential()

model2.add(Flatten(input_shape=(28, 28)))

encoder_weights = autoencoder2.layers[1].get_weights()

model2.add(Dense(units=100))

model2.layers[1].set_weights(encoder_weights)
model2.layers[1].trainable = False


model2.add(Dense(units=10, activation='sigmoid'))
```

```python
model2.summary()
```

```python
opt = keras.optimizers.SGD(learning_rate=0.2)

model2.compile(optimizer=opt,
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
l3=[]
```

```python
for epoch in range(0,200):
    history = model2.fit(x_train, y_train, epochs=1,batch_size =120, validation_data=(x_test, y_test))
    test_loss, test_accuracy = model2.evaluate(x_test, y_test)
    print(f"Epoch {epoch+1}:{test_accuracy * 100:.2f}%")
    l3.append(test_accuracy * 100)
```

```python
fig = plt.figure(figsize=(15,15))
plt.plot(l,label = "Softmax Network")
plt.plot(l2, label = "Naive Autoencoder")
plt.plot(l3, label = "Sparse Autoencoder")
plt.xlabel('Epochs')
plt.ylabel('Test Accuracy')
plt.legend()
plt.show()
```
